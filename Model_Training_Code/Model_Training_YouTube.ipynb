```python
from surprise import Dataset, Reader
from surprise import SVD
from surprise.model_selection import train_test_split
from surprise import accuracy
import pandas as pd
from sklearn.preprocessing import LabelEncoder

features = ['YouTube_channel_name', 'Content_category', 'Audience_country', 'Engagement_rate', 'Audience_age_range', 'Majority_audience_gender', 'Cost_per_post','channel_name']

maindata = youtuber_data[features]  

reader = Reader(rating_scale=(0, 1))  
dataset = Dataset.load_from_df(maindata[['YouTube_channel_name', 'Content_category', 'Cost_per_post']], reader)

trainset = dataset.build_full_trainset()
algo = SVD()
algo.fit(trainset)


def get_recommendations(category, budget, n=10):
    
    predictions = {}
    for uid in trainset.all_users():
        iid = category
        pred = algo.predict(uid, iid)
        predictions[uid] = pred.est

    sorted_predictions = sorted(predictions.items(), key=lambda x: x[1], reverse=True)

    suitable_youtubers = []
    for uid, rating in sorted_predictions:
        youtuber_info = maindata[maindata['YouTube_channel_name'] == uid].iloc[0]  
        if youtuber_info['Cost_per_post'] <= budget:
            suitable_youtubers.append((youtuber_info['YouTube_channel_name'], rating))
            if len(suitable_youtubers) == n:
                break

    return suitable_youtubers

category = 2
budget = 80000  
recommendations = get_recommendations(category, budget)
print("Recommended youtubers:")
for youtuber, rating in recommendations:
    print(f"youtuber: {youtuber}, Predicted Rating: {rating}")
```

```python
# Function to get recommendations with additional filtering by audience age range
def get_recommendations_with_age_range_filter(category, budget, min_engagement_rate, target_gender, target_country, target_age_range, n=10):
    
    predictions = {}
    for uid in trainset.all_users():
        iid = category
        pred = algo.predict(uid, iid)
        predictions[uid] = pred.est

    sorted_predictions = sorted(predictions.items(), key=lambda x: x[1], reverse=True)

    suitable_youtubers = []
    for uid, rating in sorted_predictions:
        youtuber_info = maindata[maindata['YouTube_channel_name'] == uid].iloc[0]  
        if youtuber_info['Cost_per_post'] <= budget:
            suitable_youtubers.append((youtuber_info['YouTube_channel_name'], rating))

    filtered_youtubers = []
    for youtuber, rating in suitable_youtubers:
        youtuber_info = maindata[maindata['YouTube_channel_name'] == youtuber].iloc[0]  
        if youtuber_info['Engagement_rate'] >= min_engagement_rate \
                and youtuber_info['Majority_audience_gender'] == target_gender \
                and youtuber_info['Audience_country'] == target_country \
                and youtuber_info['Audience_age_range'] == target_age_range:
            filtered_youtubers.append((youtuber, rating))
        if len(filtered_youtubers) == n:
            break

    return filtered_youtubers


category = 2
budget = 200000  
min_engagement_rate = 0.05  
target_gender = 1 
target_country = 0  
target_age_range = 3  
recommendations = get_recommendations_with_age_range_filter(category, budget, min_engagement_rate, target_gender, target_country, target_age_range)
print("Recommended youtubers:")
for youtuber, rating in recommendations:
    print(f"youtuber: {youtuber}, Predicted Rating: {rating}")
```

```python
encoded_to_id_mapping = {encoded_username: youtuber_id for encoded_username, youtuber_id in zip(maindata['YouTube_channel_name'], maindata['channel_name'])}

category = 2
budget = 200000 
min_engagement_rate =  0.05  
target_gender = 1  
target_country = 0  
target_age_range = 3  
recommendations = get_recommendations_with_age_range_filter(category, budget, min_engagement_rate, target_gender, target_country, target_age_range)
print("Recommended youtubers:")
for youtuber, rating in recommendations:
    original_youtuber_id = encoded_to_id_mapping[youtuber]
    print(f"youtuber: {original_youtuber_id}, Predicted Rating: {rating}")
```

```python
from surprise import Dataset, Reader
from surprise import SVD
from surprise.model_selection import train_test_split
from surprise import accuracy
import pandas as pd
from sklearn.preprocessing import LabelEncoder

features = ['YouTube_channel_name', 'Content_category', 'Audience_country', 'Engagement_rate', 'Audience_age_range', 'Majority_audience_gender', 'Cost_per_post','channel_name']

maindata = youtuber_data[features]  

reader = Reader(rating_scale=(0, 1))  

dataset = Dataset.load_from_df(maindata[['YouTube_channel_name', 'Content_category', 'Cost_per_post']], reader)

trainset1, testset1 = train_test_split(dataset, test_size=0.2)  
```

```python
algo = SVD()
algo.fit(trainset1)
```

```python
predictions = algo.test(testset1)
```

```python
from collections import defaultdict

def precision_recall_ndcg(predictions, k):
    ground_truth = defaultdict(list)
    for uid, iid, true_r, est, _ in predictions:
        ground_truth[uid].append((iid, true_r))

    precision_sum = 0
    recall_sum = 0
    ndcg_sum = 0
    for uid, user_ratings in ground_truth.items():
        user_ratings.sort(key=lambda x: x[1], reverse=True)
        top_k_pred = [iid for (iid, _) in user_ratings[:k]]
        
        relevant_items = sum((iid in top_k_pred) for (iid, _) in user_ratings)
        precision = relevant_items / min(k, len(user_ratings))
        recall = relevant_items / len(user_ratings)

        dcg = sum((true_r / np.log2(i + 2)) for i, (iid, true_r) in enumerate(user_ratings[:k]))
        idcg = sum((true_r / np.log2(i + 2)) for i, (iid, true_r) in enumerate(sorted(user_ratings, key=lambda x: x[1], reverse=True)[:k]))
        ndcg = dcg / idcg if idcg != 0 else 0

        precision_sum += precision
        recall_sum += recall
        ndcg_sum += ndcg

    avg_precision = precision_sum / len(ground_truth)
    avg_recall = recall_sum / len(ground_truth)
    avg_ndcg = ndcg_sum / len(ground_truth)

    return avg_precision, avg_recall, avg_ndcg

test_predictions = algo.test(testset1)

precision, recall, ndcg = precision_recall_ndcg(test_predictions, k=20)
print("Precision@20:", precision)
print("Recall@20:", recall)
print("NDCG@20:", ndcg)
```

```python
import numpy as np
from collections import defaultdict

def mean_average_precision(predictions, k):

    ground_truth = defaultdict(list)
    for uid, iid, true_r, est, _ in predictions:
        ground_truth[uid].append((iid, true_r))

    map_sum = 0
    num_users = len(ground_truth)
    for user_ratings in ground_truth.values():
        user_ratings.sort(key=lambda x: x[1], reverse=True)
        top_k_pred = [iid for iid, _ in user_ratings[:k]]
        
        ap = 0
        num_relevant = 0
        for i, (iid, true_r) in enumerate(user_ratings):
            if iid in top_k_pred:
                num_relevant += 1
                ap += num_relevant / (i + 1)
        if num_relevant > 0:
            ap /= num_relevant

        map_sum += ap

    avg_map = map_sum / num_users

    return avg_map

test_predictions = algo.test(testset1)

map = mean_average_precision(test_predictions, k=5)
print(f"Mean Average Precision@5: {map:.4f}")
```

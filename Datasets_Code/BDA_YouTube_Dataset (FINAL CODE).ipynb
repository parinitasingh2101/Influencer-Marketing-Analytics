```python
#yt1

import csv
import random
from faker import Faker

fake = Faker()

country_list = ["Brazil", "China", "Ireland", "Sweden", "Belgium", "Sweden", "Austria", "Denmark", "Netherlands",
                "United Arab Emirates", "Norway", "New Zealand", "Denmark", "Saudi Arabia", "Pakistan",
                "India", "United States", "Indonesia", "Japan", "United Kingdom", "Mexico", "France",
                "Germany", "Italy", "Argentina", "Spain", "Russia", "South Korea", "Philippines", "Colombia",
                "Thailand", "Canada", "Australia", "Egypt", "Iraq", "Malaysia", "Chile", "Poland",
                "Vietnam", "Morocco", "Peru", "Algeria", "South Africa", "Portugal", "Romania",
                "Greece", "Bangladesh", "Switzerland", "Hong Kong", "Singapore", "Costa Rica"]

gender_list = ["Male", "Female"]

columns = ["YouTube_channel_name", "Influencer_insta_username", "Content_category",
           "Audience_country", "Influencer_location",
           "Audience_age_range", "Engagement_rate", "Majority_audience_gender",
           "Cost_per_post", "Avg_likes_partnership_posts", "YouTube_subscribers",
           "YouTube_views", "YouTube_video_count"]

def generate_fake_row(insta_username, content_category, influencer_location):
    
    audience_country = random.choice(country_list) 
    youtube_subscribers = random.randint(1000, 1000000)

    min_age = random.randint(1, 19) * 5  
    max_age = min_age + 5 
    audience_age_range = f"{min_age}-{max_age}"  


    majority_audience_gender = random.choice(gender_list)

    if youtube_subscribers >= 1000 and youtube_subscribers < 10000:
        cost_per_post = "10000"
    elif youtube_subscribers >= 10000 and youtube_subscribers < 50000:
        cost_per_post = "50000"
    elif youtube_subscribers >= 50000 and youtube_subscribers < 500000:
        cost_per_post = "200000"
    elif youtube_subscribers >= 500000 and youtube_subscribers < 1000000:
        cost_per_post = "500000"
    else:
        cost_per_post = "800000"

    engagement_rate = round(random.uniform(0.5, 6.5), 1)

    if engagement_rate < 1:
        avg_likes_partnership_posts = random.randint(500, 1000)
    elif 1 <= engagement_rate <= 3.5:
        avg_likes_partnership_posts = random.randint(1000, 5000)
    elif 3.5 < engagement_rate <= 6:
        avg_likes_partnership_posts = random.randint(5000, 10000)
    else:
        avg_likes_partnership_posts = random.randint(10000, 50000)

    youtube_channel_name = fake.name()

    if engagement_rate < 1:
        youtube_views = random.randint(2000, 15000)
    elif 1 <= engagement_rate <= 3.5:
        youtube_views = random.randint(15000, 50000)
    elif 3.5 < engagement_rate <= 6:
        youtube_views = random.randint(50000, 1000000)
    else:
        youtube_views = random.randint(1000000, 10000000)

    youtube_video_count = random.randint(10, 500)

    return [youtube_channel_name, insta_username, content_category, audience_country, influencer_location,
            audience_age_range, engagement_rate, majority_audience_gender,
            cost_per_post, avg_likes_partnership_posts, youtube_subscribers,
            youtube_views, youtube_video_count]

with open("bda_data1.csv", newline="", encoding="utf-8") as csvfile:
    reader = csv.DictReader(csvfile)
    data = []
    for row in reader:
        if row["Youtuber"] == "Yes":
            insta_username = row["Influencer_insta_username"]
            content_category = row["Content_category"]
            influencer_location = row["Influencer_location"]
            data.append(generate_fake_row(insta_username, content_category, influencer_location))

with open("yt__1.csv", "w", newline="", encoding="utf-8") as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(columns)
    writer.writerows(data)

print("CSV file generated successfully.")
```
